```python
import numpy as np
import matplotlib.pyplot as plt

def plot_efficiency_curves(pump_max, gen_max):
    """
    Generates and plots the pump and generation efficiency curves as a function of load percentage.

    Args:
        pump_max (float): Maximum pumping power.
        gen_max (float): Maximum generation power.
    """
    def pump_efficiency_new(p):
        """
        Approximates pump efficiency based on power output.
        """
        if pump_max == 0:
            return 0.0
        p_norm = p / pump_max
        efficiency = 0.8 + 0.1 * (1 - (p_norm - 0.85)**2 / 0.03)
        return np.clip(efficiency, 0, 0.9)

    def gen_efficiency_new(g):
        """
        Approximates generator efficiency based on power output.
        """
        if gen_max == 0:
            return 0.0
        g_norm = g / gen_max
        efficiency = 0.55 + 0.3 * (1 - (g_norm - 0.65)**2 / 1.5)
        return np.clip(efficiency, 0.7, 0.95)

    # Generate power values for plotting
    p_powers = np.linspace(0, pump_max, 100)
    g_powers = np.linspace(0, gen_max, 100)

    # Calculate efficiencies
    pump_efficiencies = [pump_efficiency_new(p) for p in p_powers]
    gen_efficiencies = [gen_efficiency_new(g) for g in g_powers]

    # Calculate load percentages
    pump_load_percentage = (p_powers / pump_max) * 100 if pump_max else np.zeros_like(p_powers)
    gen_load_percentage = (g_powers / gen_max) * 100 if gen_max else np.zeros_like(g_powers)

    # Plotting
    plt.figure(figsize=(8, 6))
    plt.plot(pump_load_percentage, pump_efficiencies, label="Pumping Efficiency", color='r')
    plt.plot(gen_load_percentage, gen_efficiencies, label="Generation Efficiency", color='g')
    plt.xlabel("Load (%)")
    plt.ylabel("Efficiency")
    plt.title("Efficiency vs. Load")
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == '__main__':
    # Example usage:
    plot_efficiency_curves(pump_max=100, gen_max=100)  # Replace with your actual max power values
```
